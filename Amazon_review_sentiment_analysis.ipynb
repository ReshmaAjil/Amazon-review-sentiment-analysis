```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("mohamedbakhet/amazon-books-reviews")

print("Path to dataset files:", path)
```

```python
import os

os.listdir(path)
```

```python
import pandas as pd

# Load both files
data = pd.read_csv(path + "/books_data.csv")
rating = pd.read_csv(path + "/Books_rating.csv")

# Display first rows
data.head(), rating.head()
```

```python
# Merge both datasets on Title
merged = pd.merge(rating, data, on="Title", how="inner")

merged.head()
merged.info()
```

```python
# Function to assign sentiment category based on rating score
def sentiment_label(score):
    if score >= 4.0:
        return "Positive"
    elif score <= 2.0:
        return "Negative"
    else:
        return "Neutral"

merged["sentiment"] = merged["review/score"].apply(sentiment_label)

merged["sentiment"].value_counts()
```

```python
# Select a random sample of 200,000 rows from the merged dataset
sample = merged.sample(200000, random_state=42)   # 200k random rows
sample.shape
```

```python
import re
import string
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download("stopwords")
nltk.download("wordnet")

# Creating stopword list and lemmatizer object
stop_words = set(stopwords.words('english'))
# Create a set of english stopwords
lemmatizer = WordNetLemmatizer()

def clean_text(text):
    text = text.lower()
    text = re.sub(r"http\S+|www\S+|https\S+","",text)
    text = re.sub(r"[^\w\s]","",text)
    text = re.sub(r"\d+","",text)
    words = text.split()
    words = [w for w in words if w not in stop_words]
    words = [lemmatizer.lemmatize(w) for w in words]
    return " ".join(words)

sample["cleaned_text"] = sample["review/text"].astype(str).apply(clean_text)
sample.head()
```
